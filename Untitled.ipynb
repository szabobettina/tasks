```python
def add_row_to_order_items_table(last_order_id=None):
    def submit_data():
        order_id = orderid_entry.get()
        product_id = productid_entry.get()
        quantity = quantity_entry.get()

        if not order_id or not product_id or not quantity:
            messagebox.showerror("Hiba", "A rendelés azonosítója, termék azonosítója és mennyiség megadása kötelező!")
            return

        if not is_int(quantity) or int(quantity) < 1:
            messagebox.showerror("Hiba", "Az a rendelt mennyiség érvénytelen!")
            return

        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            cursor.execute("SELECT * FROM Orders WHERE id = ?", (order_id,))
            order_exists = cursor.fetchone()

            cursor.execute("SELECT * FROM Products WHERE id = ?", (product_id,))
            product_exists = cursor.fetchone()

            if not order_exists:
                messagebox.showerror("Hiba", "A megadott rendelés azonosító nem létezik!")
                return

            if not product_exists:
                messagebox.showerror("Hiba", "A megadott termék azonosító nem létezik!")
                return

            cursor.execute("SELECT * FROM Order_items WHERE order_id = ? AND product_id = ?", (order_id, product_id))
            existing_order_item = cursor.fetchone()

            cursor.execute("SELECT StockQuantity FROM Products WHERE id = ?", (product_id,))
            stock_quantity = cursor.fetchone()[0]

            if int(quantity) > stock_quantity:
                messagebox.showerror("Hiba", "A rendelt mennyiség nem lehet nagyobb, mint a rendelkezésre álló raktárkészlet!")
                return

            if existing_order_item:
                updated_quantity = int(existing_order_item[3]) + int(quantity)
                cursor.execute("UPDATE Order_items SET quantity = ? WHERE order_id = ? AND product_id = ?", (updated_quantity, order_id, product_id))
            else:
                cursor.execute("INSERT INTO Order_items (order_id, product_id, quantity) VALUES (?, ?, ?)", (order_id, product_id, quantity))

            cursor.execute("SELECT Price FROM Products WHERE id = ?", (product_id,))
            price = cursor.fetchone()[0]

            cursor.execute("SELECT grand_total FROM Orders WHERE id = ?", (order_id,))
            current_total = cursor.fetchone()[0]

            new_total = current_total + (price * int(quantity))
            cursor.execute("UPDATE Orders SET grand_total = ? WHERE id = ?", (new_total, order_id))

            updated_stock_quantity = stock_quantity - int(quantity)
            cursor.execute("UPDATE Products SET StockQuantity = ? WHERE id = ?", (updated_stock_quantity, product_id))

            conn.commit()
            conn.close()
            messagebox.showinfo("Siker", "Rendelési tétel sikeresen hozzáadva vagy frissítve!")
            add_window.destroy()

        except sqlite3.Error as error:
            messagebox.showerror("Hiba", str(error))

    add_window = tk.Toplevel(root)
    add_window.title("Új rendelési tétel hozzáadása")
    add_window.geometry("400x400")

    orderid_label = tk.Label(add_window, text="Rendelés azonosítója:")
    orderid_label.pack(pady=5)
    orderid_entry = tk.Entry(add_window)
    if last_order_id is not None:
        orderid_entry.insert(0, last_order_id)  # automatikus rendelésazonosító, ha van ilyen
    orderid_entry.pack(pady=5)

    productid_label = tk.Label(add_window, text="Termék azonosítója:")
    productid_label.pack(pady=5)
    productid_entry = tk.Entry(add_window)
    productid_entry.pack(pady=5)

    quantity_label = tk.Label(add_window, text="Mennyiség:")
    quantity_label.pack(pady=5)
    quantity_entry = tk.Entry(add_window)
    quantity_entry.pack(pady=5)

    submit_button = tk.Button(add_window, text="Hozzáadás/Frissítés", command=submit_data)
    submit_button.pack(pady=20)
```

```python
import sqlite3
import tkinter as tk
from tkinter import ttk
from tkinter import simpledialog, messagebox, Menu
import hashlib
import pandas as pd
from tkinter import filedialog
from IPython.display import clear_output
from matplotlib.figure import Figure 
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt

conn = sqlite3.connect('users.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    username TEXT UNIQUE NOT NULL,
    password TEXT NOT NULL
)
''')
conn.commit()
conn.close()

def perform_registration(username, password, confirm_password, register_window):
    if not username or not password or not confirm_password:
        messagebox.showerror("Hiba", "A mezők kitöltése kötelező!")
        register_window.destroy()
        return

    if password != confirm_password:
        messagebox.showerror("Hiba", "A jelszavak nem egyeznek!")
        register_window.destroy()
        return

    if register_user(username, password):
        messagebox.showinfo("Siker", "Sikeres regisztráció!")
        register_window.destroy()
        
    else:
        messagebox.showerror("Hiba", "Ez a felhasználónév már foglalt!")
        register_window.destroy()
    

def register_user(username, password):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    try:
        cursor.execute("INSERT INTO users (username, password) VALUES (?, ?)", (username, hashed_password))
        conn.commit()
        conn.close()
        return True
    except sqlite3.IntegrityError:
        return False

def register():
    register_window = tk.Toplevel(root)
    register_window.title("Regisztráció")
    register_window.geometry("300x300")
    
    reg_username_label = tk.Label(register_window, text="Felhasználónév:")
    reg_username_label.pack(pady=5)
    reg_username_entry = tk.Entry(register_window)
    reg_username_entry.pack(pady=5)
    
    reg_password_label = tk.Label(register_window, text="Jelszó:")
    reg_password_label.pack(pady=5)
    reg_password_entry = tk.Entry(register_window, show="*")
    reg_password_entry.pack(pady=5)
    
    reg_confirm_password_label = tk.Label(register_window, text="Jelszó megerősítése:")
    reg_confirm_password_label.pack(pady=5)
    reg_confirm_password_entry = tk.Entry(register_window, show="*")
    reg_confirm_password_entry.pack(pady=5)
    
    def on_register_button_clicked():
        perform_registration(
            reg_username_entry.get(), 
            reg_password_entry.get(), 
            reg_confirm_password_entry.get(),
            register_window)

    reg_submit_button = tk.Button(register_window, text="Regisztráció", command=on_register_button_clicked)
    reg_submit_button.pack(pady=20)


def add_row_to_table(table_name):
    print(table_name)
    if table_name == "Customers":
        add_row_to_customers_table()
    if table_name == "Products":
        add_row_to_products_table()
    if table_name == "Order_items":
        add_row_to_order_items_table()
    if table_name == "Orders":
        add_row_to_orders_table()
        
        
def delete_row_from_table(table_name):
    print("ok! 130")
    def delete_data():
        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            if not row_exists(cursor, row_id, table_name):
                messagebox.showerror("Hiba", f"Nincs ilyen azonosítójú sor a táblában: {row_id}")
                conn.close()
                delete_window.destroy()
                return
            
            confirmation = messagebox.askyesno("Megerősítés", f"Biztosan törli a {row_id} azonosítójú sort?")
            if not confirmation:
                conn.close()
                delete_window.destroy()
                return
            
            if has_orders(cursor, row_id, table_name):
                if table_name == "Products":
                    messagebox.showwarning("Figyelmeztetés", "Ez a termék rendelési tételekhez vagy rendelésekhez van kapcsolva. "
                                                              "A törlés esetén a készletmennyiséget 0-ra állítjuk.")
                    cursor.execute("UPDATE Products SET StockQuantity = 0 WHERE id = ?", (row_id,))
                else:
                    delete_order_item_and_update_tables(table_name, row_id)
                    
            else:
                cursor.execute(f"DELETE FROM {table_name} WHERE id = ?", (row_id,))


                
            conn.commit()
            conn.close()

            messagebox.showinfo("Siker", "A sor sikeresen törölve lett!")
            delete_window.destroy()

        except sqlite3.Error as error:
            messagebox.showerror("Hiba", str(error))
            
    def delete_order_item_and_update_tables(table_name, selected_row_id):
        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            if table_name == "Order_items":
                cursor.execute("SELECT order_id, product_id, quantity FROM Order_items WHERE id = ?", (selected_row_id,))
                order_item = cursor.fetchone()

                if not order_item:
                    print("Nincs ilyen rendelési tétel az adatbázisban.")
                    return

                order_id, product_id, quantity = order_item
                
                cursor.execute("SELECT Price FROM Products WHERE id = ?", (product_id,))
                price = cursor.fetchone()[0]

                cursor.execute("DELETE FROM Order_items WHERE id = ?", (selected_row_id,))

                cursor.execute("UPDATE Orders SET grand_total = grand_total - ? WHERE id = ?", (price * quantity, order_id))

                cursor.execute("UPDATE Products SET StockQuantity = StockQuantity + ? WHERE id = ?", (quantity, product_id))

                conn.commit()
                
            elif table_name == "Orders":
                print("orders 204 ok")
                cursor.execute("SELECT id, product_id, quantity FROM Order_items WHERE order_id = ?", (selected_row_id,))
                order_item = cursor.fetchall()
                
                cursor.execute("DELETE FROM Orders WHERE id = ?", (selected_row_id,))

                for item_id, product_id, quantity, in order_item:                    
                    cursor.execute("UPDATE Products SET StockQuantity = StockQuantity + ? WHERE id = ?", (quantity, product_id))
                    
                    cursor.execute("DELETE FROM Order_items WHERE id = ?", (item_id,))
                    
                    print(item_id, product_id, quantity)
                    
                conn.commit()                   
                    

        except sqlite3.Error as error:
            print("Hiba a delete_order_item_and_update_tables függvényben:", str(error))

            
    def has_orders(cursor, product_id, table_name):
        if table_name == "Products":
            cursor.execute("SELECT id FROM Order_items WHERE product_id = ? LIMIT 1", (product_id,))
            order_item = cursor.fetchone()

            cursor.execute("SELECT id FROM Orders WHERE id IN (SELECT order_id FROM Order_items WHERE product_id = ?) LIMIT 1", (product_id,))
            order = cursor.fetchone()

            return order_item is not None or order is not None
        elif table_name == "Order_items":
            cursor.execute("SELECT id FROM Orders WHERE id = (SELECT order_id FROM Order_items WHERE id = ?) LIMIT 1", (product_id,))
            return cursor.fetchone() is not None
        else:
            cursor.execute("SELECT id FROM Order_items WHERE order_id = ? LIMIT 1", (product_id,))
            return cursor.fetchone() is not None

        return False

    def row_exists(cursor, item_id, table_name):
        try:
            cursor.execute(f"SELECT id FROM {table_name} WHERE id = ?", (item_id,))
            return cursor.fetchone() is not None
        except sqlite3.Error as error:
            print("Hiba a row_exists függvényben:", str(error))
            return False


    delete_window = tk.Toplevel()
    delete_window.title(f"Sor törlése - {table_name}")
    delete_window.geometry("300x100")

    row_id = simpledialog.askinteger("Sor törlése", "Kérem, adja meg a törlendő sor azonosítóját:")
    if not row_id:
        return

    delete_button = tk.Button(delete_window, text="Törlés", command=delete_data)
    delete_button.pack(pady=20)
     
def validate_date_format(date):
    try:
        datetime.strptime(date, "%Y-%m-%d")
        return True
    except ValueError:
        return False

def edit_row_in_table(table_name):
    def edit_data():
        nonlocal edited_data
        edited_data = {}  # Új adatokat tartalmazó szótár
        
        for key in column_names:
            # Az id és grand_total mezőket nem szerkeszthetjük
            if key != "id" and (table_name != "Orders" or key != "grand_total"):
                if key == "StockQuantity" and table_name == "Products":
                    try:
                        stock_quantity_value = int(entry_vars[key].get())
                    except ValueError:
                        messagebox.showerror("Hiba", f"{key} nem megfelelő!")
                        return
                    if stock_quantity_value < 0:
                        messagebox.showerror("Hiba", f"{key} nem lehet kisebb, mint 0!")
                        return
                    edited_data[key] = stock_quantity_value
                elif key == "Price" and table_name == "Products":
                    try:
                        price_value = float(entry_vars[key].get())
                    except ValueError:
                        messagebox.showerror("Hiba", f"{key} nem megfelelő!")
                        return
                    if price_value < 1:
                        messagebox.showerror("Hiba", f"{key} nem lehet kisebb, mint 1!")
                        return
                    edited_data[key] = price_value
                elif key == "quantity" and table_name == "Order_items":
                    try:
                        quantity_value = int(entry_vars[key].get())
                    except ValueError:
                        messagebox.showerror("Hiba", f"{key} nem megfelelő!")
                        return
                    
                    product_id = entry_vars["product_id"].get()
                    
                    if (quantity_value < 1) or (quantity_value > get_stock_quantity(product_id)):
                        messagebox.showerror("Hiba", "Nem megfelelő mennyiség!")
                        return
                    
                    old_order_quantity = get_order_item_quantity(row_id)
                    new_order_quantity = quantity_value
                    
                    print(old_order_quantity, new_order_quantity)
                    
                    update_stock_quantity(product_id, old_order_quantity, new_order_quantity)
                    edited_data[key] = quantity_value

                else:
                    edited_data[key] = entry_vars[key].get()

        if any(value == "" for value in edited_data.values()):
            messagebox.showerror("Hiba", "Minden mező kitöltése kötelező!")
            return

        if table_name == "Orders" and not validate_date_format(edited_data.get("order_date", "")):
            messagebox.showerror("Hiba", "Helytelen dátum formátum! (Pl.: 2023-10-26)")
            return
        
        if table_name == "Order_items":
            order_id = edited_data.get("order_id")
            product_id = edited_data.get("product_id")
            if not order_id or not product_id:
                messagebox.showerror("Hiba", "Az order_id és product_id mezők kitöltése kötelező!")
                return

            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            # Ellenőrizzük, hogy a rendelés létezik
            if not order_exists(cursor, order_id, "Orders"):
                messagebox.showerror("Hiba", f"Nincs ilyen rendelés azonosító ({order_id}) az adatbázisban.")
                conn.close()  # Ne felejtsük el lezárni az adatbázis kapcsolatot
                return

            # Ellenőrizzük, hogy a termék létezik
            if not order_exists(cursor, product_id, "Products"):
                messagebox.showerror("Hiba", f"Nincs ilyen termék azonosító ({product_id}) az adatbázisban.")
                conn.close()
                return

            conn.close()

        edit_row(edited_data)

    def get_order_item_quantity(order_item_id):
        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            cursor.execute("SELECT quantity FROM Order_items WHERE id = ?", (order_item_id,))
            quantity = cursor.fetchone()[0]

            conn.close()

            return quantity

        except sqlite3.Error as error:
            print("Hiba a get_order_item_quantity függvényben:", str(error))
            return 0


    def edit_row(data):
        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            # Az új adatokat az SQL UPDATE parancshoz hozzáadjuk
            update_values = ", ".join([f"{key} = ?" for key in data.keys()])
            update_data = tuple(data.values())

            cursor.execute(f"UPDATE {table_name} SET {update_values} WHERE id = ?", (*update_data, row_id))

            if table_name == "Order_items":
                print(row_id)
                update_grand_total_order_items(cursor, row_id)
                
            if table_name == "Products":
                update_grand_total_products(cursor, row_id)
                

            conn.commit()
            conn.close()

            messagebox.showinfo("Siker", "A sor sikeresen frissítve lett!")
            edit_window.destroy()

        except sqlite3.Error as error:
            messagebox.showerror("Hiba", str(error))
            
    def update_grand_total_order_items(cursor, order_items_id):
        try:
            cursor.execute("SELECT order_id, product_id, quantity FROM Order_items WHERE id = ?", (order_items_id,))
            order_item = cursor.fetchone()

            if not order_item:
                print("Nincs ilyen rendelési tétel az adatbázisban.")
                return

            order_id, product_id, new_quantity = order_item

            #termék ára
            cursor.execute("SELECT Price FROM Products WHERE id = ?", (product_id,))
            price = cursor.fetchone()[0]

            #összes rendelési tétel a megadott rendeléshez
            cursor.execute("SELECT quantity FROM Order_items WHERE order_id = ?", (order_id,))
            order_items = cursor.fetchall()

            total = 0

            for quantity, in order_items:
                total += price * quantity

            # Frissítjük a rendeléshez tartozó grand_total értéket
            cursor.execute("UPDATE Orders SET grand_total = ? WHERE id = ?", (total, order_id))

        except sqlite3.Error as error:
            print("Hiba az update_grand_total függvényben:", str(error))
            
    def update_grand_total_products(cursor, product_id):
        try:
            # termék ára
            cursor.execute("SELECT Price FROM Products WHERE id = ?", (product_id,))
            new_price = cursor.fetchone()[0]

            #összes rendelési tétel a megadott termékhez
            cursor.execute("SELECT order_id, quantity FROM order_items WHERE product_id = ?", (product_id,))
            order_items = cursor.fetchall()

            for order_id, quantity in order_items:
                # Frissítjük az összértéket a módosított ár alapján
                total = new_price * quantity
                cursor.execute("UPDATE Orders SET grand_total = ? WHERE id = ?", (total, order_id))

        except sqlite3.Error as error:
            print("Hiba az update_grand_total_products függvényben:", str(error))

    def load_data():
        try:
            conn = sqlite3.connect('sqliteOrder.db')
            cursor = conn.cursor()

            #szerkeszteni kívánt sor
            cursor.execute(f"SELECT * FROM {table_name} WHERE id = ?", (row_id,))
            row_data = cursor.fetchone()

            conn.close()
            
            if not row_data:
                messagebox.showerror("Hiba", f"Nincs ilyen azonosítójú sor a táblában: {row_id}")
                edit_window.destroy()
                return None

            return row_data

        except sqlite3.Error as error:
            messagebox.showerror("Hiba", str(error))
            return None

    row_id = simpledialog.askinteger("Sor azonosítója", "Kérem, adja meg a szerkeszteni kívánt sor azonosítóját:")
    if not row_id:
        return

    edited_data = {}  # új adatokat

    try:
        conn = sqlite3.connect('sqliteOrder.db')
        cursor = conn.cursor()

        #tábla oszlopnevei
        cursor.execute(f"PRAGMA table_info({table_name})")
        columns = cursor.fetchall()
        column_names = [column[1] for column in columns]

        conn.close()

    except sqlite3.Error as error:
        messagebox.showerror("Hiba", str(error))
        return

    edit_window = tk.Toplevel()
    edit_window.title(f"Sor szerkesztése - {table_name}")
    edit_window.geometry("400x300")

    # Adatok betöltése a szerkesztő ablakba
    row_data = load_data()

    if not row_data:
        return

    entry_vars = {}  # Ebben a változóban tároljuk a beviteli mezőkhöz tartozó változókat

    for i, key in enumerate(column_names):
        label = tk.Label(edit_window, text=key)
        label.grid(row=i, column=0, padx=5, pady=5)

        entry_var = tk.StringVar(value=row_data[i])  # Az alapértelmezett érték a jelenlegi adat
        entry = tk.Entry(edit_window, textvariable=entry_var, state='disabled' if key == "id" or key == "grand_total" or key == "product_id" or key == "order_id" else 'normal')
        entry.grid(row=i, column=1, padx=5, pady=5)

        entry_vars[key] = entry_var

    edit_button = tk.Button(edit_window, text="Szerkesztés", command=edit_data)
    edit_button.grid(row=len(column_names), column=0, columnspan=2, pady=10)
    
def order_exists(cursor, item_id, table_name):
    try:
        cursor.execute(f"SELECT id FROM {table_name} WHERE id = ?", (item_id,))
        return cursor.fetchone() is not None
    except sqlite3.Error as error:
        print("Hiba az order_exists függvényben:", str(error))
        return False
    
def update_stock_quantity(product_id, old_order_quantity, new_order_quantity):
    try:
        conn = sqlite3.connect('sqliteOrder.db')
        cursor = conn.cursor()
        cursor.execute("SELECT StockQuantity FROM Products WHERE id = ?", (product_id,))
        original_stock_quantity = cursor.fetchone()[0]

        new_stock_quantity = original_stock_quantity - (new_order_quantity - old_order_quantity)

        cursor.execute("UPDATE Products SET StockQuantity = ? WHERE id = ?", (new_stock_quantity, product_id))
        
        conn.commit()
        conn.close()

    except sqlite3.Error as error:
        print("Hiba az update_stock_quantity függvényben:", str(error))

        
def get_stock_quantity(product_id):
    try:
        conn = sqlite3.connect('sqliteOrder.db')
        cursor = conn.cursor()
        
        cursor.execute("SELECT StockQuantity FROM Products WHERE id = ?", (product_id,))
        stock_quantity = cursor.fetchone()[0]

        conn.close()

        return stock_quantity
    
    except sqlite3.Error as error:
        print("Hiba a get_stock_quantity függvényben:", str(error))
        return 0

def show_related_rows(row_data, table_name):
    related_rows = None
    if table_name == "Orders":
        data_id = row_data[0] 
        query = "SELECT * FROM Order_items WHERE order_id = ?"
        title = "Rendelés részletei"        
    if table_name == "Order_items":
        data_id = row_data[2] 
        query = "SELECT * FROM Products WHERE id = ?"
        title = "Termék részletek"
        
    conn = sqlite3.connect('sqliteOrder.db')
    cursor = conn.cursor()
    cursor.execute(query, (data_id,))
    related_rows = cursor.fetchall()
    column_names = [description[0] for description in cursor.description]
    conn.close()

    if related_rows:
        new_window = tk.Toplevel()
        new_window.title(title)
        new_window.geometry("500x400")

        tree = ttk.Treeview(new_window, columns=column_names, show="headings")

        for name in column_names:
            tree.heading(name, text=name)
            tree.column(name, width=30)


        for row in related_rows:
            tree.insert("", "end", values=row)

        tree.pack(expand=True, fill="both")
    elif not related_rows and table_name == "Orders":
        messagebox.showinfo("Hiba", "Ehhez a rendeléshez nincs még rendelési tétel!")
        return
        
    
    
    def on_select(event):
        item = tree.selection()[0]  # Kijelölt elem
        if item:
            selected_row = tree.item(item, "values")
            show_related_rows(selected_row, "Order_items")
    if title == "Rendelés részletei":
        tree.bind("<<TreeviewSelect>>", on_select)



def list_rows_in_table(table_name):
    # Új ablak létrehozása
    list_window = tk.Toplevel(root)
    list_window.title(f"{table_name} tartalma")
    list_window.geometry("600x400")

    conn = sqlite3.connect('sqliteOrder.db')
    cursor = conn.cursor()

    cursor.execute(f"SELECT * FROM {table_name}")
    rows = cursor.fetchall()

    # Tábla oszlopainak nevei
    column_names = [description[0] for description in cursor.description]

    conn.close()

    tree = ttk.Treeview(list_window, columns=column_names, show="headings")

    for name in column_names:
        tree.heading(name, text=name)
        tree.column(name, width=30)

    for row in rows:
        tree.insert("", "end", values=row)

    tree.pack(expand=True, fill="both")

    def on_select(event):
        item = tree.selection()[0]  # Kijelölt elem
        if item:
            selected_row = tree.item(item, "values")
            show_related_rows(selected_row, table_name)

    if table_name == "Orders" or table_name == "Order_items":
        tree.bind("<<TreeviewSelect>>", on_select)

def export_rows_from_table(table_name):
    conn = sqlite3.connect('sqliteOrder.db')
    cursor = conn.cursor()

    cursor.execute(f"SELECT * FROM {table_name}")
    rows = cursor.fetchall()

    column_names = [description[0] for description in cursor.description]

    conn.close()

    # DataFrame létrehozása a pandas segítségével
    df = pd.DataFrame(rows, columns=column_names)

    file_path = filedialog.asksaveasfilename(defaultextension=".xlsx", filetypes=[("Excel files", "*.xlsx")])

    if file_path:
        df.to_excel(file_path, index=False)

def import_rows_to_table(table_name):
    file_path = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx"), ("Text files", "*.txt")])

    if not file_path:
        return

    try:
        if file_path.endswith(".xlsx"):
            df = pd.read_excel(file_path)
        elif file_path.endswith(".txt"):
            df = pd.read_csv(file_path, delimiter="\t")  # Például tabulátumokkal elválasztott szövegfájl
        else:
            messagebox.showerror("Hiba", "Nem támogatott fájlformátum.")
            return

        if validate_and_insert_data(df, table_name):
            messagebox.showinfo("Siker", "A sikeres importálás!")
        else:
            messagebox.showwarning("Figyelem", f"Nem sikerült importálni a(z) {table_name} táblát.")

    except Exception as e:
        messagebox.showerror("Hiba", str(e))

def validate_and_insert_data(df, table_name):
    conn = sqlite3.connect('sqliteOrder.db')
    cursor = conn.cursor()

    if table_name == "Orders":
        for index, row in df.iterrows():
            name = row.get("name")
            grand_total = row.get("grand_total")
            order_date = row.get("order_date")

            if not name or not grand_total or not order_date:
                messagebox.showerror("Hiba", "Hiányzó adatok az importált fájlban.")
                return
            
            if not validate_date_format(order_date) or not is_float(grand_total):
                messagebox.showerror("Hiba", "Helytelen adatok az importált fájlban.")
                return

            try:
                cursor.execute("INSERT INTO Orders (name, grand_total, order_date) VALUES (?, ?, ?)",
                               (name, grand_total, order_date))
            except sqlite3.Error as error:
                messagebox.showerror("Hiba", str(error))
                return

    elif table_name == "Order_items":
        for index, row in df.iterrows():
            order_id = row.get("order_id")
            product_id = row.get("product_id")
            quantity = row.get("quantity")

            if not order_id or not product_id or not quantity:
                messagebox.showerror("Hiba", "Hiányzó adatok az importált fájlban.")
                return
            
            if not is_int(order_id) or not is_int(product_id) or not is_int(quantity):
                messagebox.showerror("Hiba", "Helytelen adatok az importált fájlban.")
                return

            try:
                cursor.execute("INSERT INTO Order_items (order_id, product_id, quantity) VALUES (?, ?, ?)",
                               (order_id, product_id, quantity))
            except sqlite3.Error as error:
                messagebox.showerror("Hiba", str(error))
                return

    elif table_name == "Products":
        for index, row in df.iterrows():
            product_name = row.get("ProductName")
            description = row.get("Description")
            price = row.get("Price")
            stock_quantity = row.get("StockQuantity")

            if not product_name or not price or not stock_quantity:
                messagebox.showerror("Hiba", "Hiányzó adatok az importált fájlban.")
                return
            if not is_float(price) or not is_int(stock_quantity):
                messagebox.showerror("Hiba", "Helytelen adatok az importált fájlban.")
                return
            
            try:
                cursor.execute("INSERT INTO Products (ProductName, Description, Price, StockQuantity) VALUES (?, ?, ?, ?)",
                               (product_name, description, price, stock_quantity))
            except sqlite3.Error as error:
                messagebox.showerror("Hiba", str(error))
                return

    conn.commit()
    conn.close()
    return True


def create_subwindow_with_menu(title):
    if title == "Products":
        table = "Termékek"
    elif title == "Order_items":
        table = "Rendelési tételek"
    else:
        table = "Rendelések"
         
    window = tk.Toplevel(root)
    window.title(table)
    window.geometry("280x180")

    menu_bar = tk.Menu(window)
    window.config(menu=menu_bar)
    
    label = tk.Label(window, text= f"Üdvözöljük a(z) {table} ablakban!")
    label.pack()

    operations_menu = tk.Menu(menu_bar, tearoff=0)
    menu_bar.add_cascade(label="Műveletek", menu=operations_menu)
    operations_menu.add_command(label="Új sor hozzáadása", command=lambda: add_row_to_table(title))
    operations_menu.add_command(label="Sor törlése", command=lambda: delete_row_from_table(title))
    operations_menu.add_command(label="Sor szerkesztése", command=lambda: edit_row_in_table(title))
    operations_menu.add_command(label="Tartalom listázása", command=lambda: list_rows_in_table(title))
    operations_menu.add_command(label="Exportálás", command=lambda: export_rows_from_table(title))
    operations_menu.add_command(label="Importálás", command=lambda: import_rows_to_table(title))


def open_customers_window():
    create_subwindow_with_menu("Customers")

def open_products_window():
    create_subwindow_with_menu("Products")

def open_order_items_window():
    create_subwindow_with_menu("Order_items")

def open_orders_window():
    create_subwindow_with_menu("Orders")
    
def check_login(username, password):
    conn = sqlite3.connect('users.db')
    cursor = conn.cursor()
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    cursor.execute("SELECT * FROM users WHERE username = ? AND password = ?", (username, hashed_password))
    user = cursor.fetchone()
    conn.close()
    return user is not None


def draw_login_ui():
    username_label.pack(pady=5)
    username_entry.pack(pady=5)
    password_label.pack(pady=5)
    password_entry.pack(pady=5)
    login_button.pack(pady=20)
    register_button.pack(pady=10)

def attempt_login():
    username = username_entry.get()
    password = password_entry.get()

    if check_login(username, password):
        for widget in root.winfo_children():
            widget.pack_forget()  # rejtse el az összes widgetet
        draw_main_menu_ui()
    else:
        messagebox.showerror("Hiba", "Érvénytelen felhasználónév vagy jelszó!")
        


def logout():
    global root
    root.withdraw()
    draw_login_ui() 
    login_button.config(state=tk.NORMAL)
    clear_output(wait=True)
    
def fetch_data(query):
    cursor.execute(query)
    data = cursor.fetchall()
    return data

def open_visualization_window(title, query, chart_type):
    window = tk.Toplevel(root)
    window.title(title)

    data = fetch_data(query)

    fig, ax = plt.subplots()

    if chart_type == 'bar':
        ax.bar([item[0] for item in data], [item[1] for item in data])
        ax.set_xlabel('Termék neve')
        ax.set_ylabel('Darabszám')
    elif chart_type == 'pie':
        ax.pie([item[1] for item in data], labels=[item[0] for item in data], autopct='%1.1f%%', startangle=90)
        ax.axis('equal')  # Az ábrát kör alakúra állítja

    ax.set_title(title)

    canvas = FigureCanvasTkAgg(fig, master=window)
    canvas.draw()
    canvas.get_tk_widget().pack()

def open_summary():
    query = '''
        SELECT ProductName, SUM(quantity) as Darabszam
        FROM Order_items
        INNER JOIN Products ON Order_items.product_id = Products.id
        GROUP BY ProductName
        ORDER BY Darabszam DESC
        LIMIT 1
    '''
    data = fetch_data(query)
    query = '''
        SELECT SUM(grand_total) as Teljes_Osszeg FROM Orders
    '''
    max_revenue = fetch_data(query)
    income = max_revenue[0][0]
    if data:
        product_name, sold_quantity = data[0]
        summary_window = tk.Tk()
        summary_window.title("Összegzés")
        label = tk.Label(summary_window, text=f"A legtöbbet eladott termék: {product_name}\n")
        label.pack(pady = 5)
        label = tk.Label(summary_window, text=f"Eladott darabszám: {sold_quantity}\n")
        label.pack(pady = 5)
        label = tk.Label(summary_window, text=f"Maximális bevétel: {income} Ft")
        label.pack(pady = 5)
        ok_button = tk.Button(summary_window, text="Ok", command=summary_window.destroy)
        ok_button.pack(pady=20)
        summary_window.geometry("200x200")
    else:
        messagebox.showinfo("Összegzés", "Nincs eladott termék az adatbázisban.")

def open_best_selling():
    query = '''
        SELECT ProductName, SUM(quantity) as Darabszam
        FROM Order_items
        INNER JOIN Products ON Order_items.product_id = Products.id
        GROUP BY ProductName
        ORDER BY Darabszam DESC
        LIMIT 5
    '''
    open_visualization_window("Legtöbbet eladott termékek", query, 'bar')

def open_largest_orders():
    query = '''
        SELECT Orders.id, grand_total
        FROM Orders
        ORDER BY grand_total DESC
    '''
    open_visualization_window("Legnagyobb összegű rendelések", query, 'pie')

        
def draw_main_menu_ui():
    username = username_entry.get()
    root.title("Főmenü")
    root.geometry("370x250")

    label = tk.Label(root, text="Üdvözöljük a Főmenüben, " + username + "!")
    label.pack(pady=20)
    
    menu_bar = tk.Menu(root)
    root.config(menu=menu_bar)

    # Gombok hozzáadása
    customers_button = tk.Button(root, text="Ügyfelek", state="disabled", command=open_customers_window)
    customers_button.pack(side="left", padx=10)
    
    #label = tk.Label(root, text="Termékek ablak megnyitása")
    #label.pack()
    products_button = tk.Button(root, text="Termékek", command=open_products_window)
    products_button.pack(side="left", padx=10)

    #label = tk.Label(root, text="Rendelés tételek ablak megnyitása")
    #label.pack()
    order_items_button = tk.Button(root, text="Rendelés tételek", command=open_order_items_window)
    order_items_button.pack(side="left", padx=10)

    #label = tk.Label(root, text="Rendelések ablak megnyitása")
    #label.pack()
    orders_button = tk.Button(root, text="Rendelések", command=open_orders_window)
    orders_button.pack(side="left", padx=10)
    
    account_menu = tk.Menu(menu_bar, tearoff=0)
    menu_bar.add_cascade(label="Fiók", menu=account_menu)
    account_menu.add_command(label="Kijelentkezés", state='disabled', command=logout)
    visualisation_menu = account_menu = tk.Menu(menu_bar, tearoff=0)
    menu_bar.add_cascade(label="Kimutatások", menu=visualisation_menu)
    account_menu.add_command(label="Összegzés", command=open_summary)
    account_menu.add_command(label="Legtöbbet eladott termékek",  command=open_best_selling)
    account_menu.add_command(label="Legnagyobb összegű rendelések", command=open_largest_orders)
    
conn = sqlite3.connect('sqliteOrder.db')
cursor = conn.cursor()
    
root = tk.Tk()
root.title("Bejelentkezés")
label = tk.Label(root, text="Üdvözöljük a Főmenüben!")
label.pack()
label = tk.Label(root, text="A folytatáshoz jelentkezzen be!")
label.pack(pady=5)
root.geometry("300x300")
username_label = tk.Label(root, text="Felhasználónév:")
username_entry = tk.Entry(root)
password_label = tk.Label(root, text="Jelszó:")
password_entry = tk.Entry(root, show="*")
login_button = tk.Button(root, text="Belépés", command=attempt_login)
register_button = tk.Button(root, text="Regisztráció", command=register)

draw_login_ui()
root.mainloop()
conn.close()
```
